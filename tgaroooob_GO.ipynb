```python
%%writefile chatbot.go
package main

import (
	"bufio"
	"context"
	"fmt"
	"log"
	"os"
	"strings"

	"github.com/google/generative-ai-go/genai"
	"google.golang.org/api/option"

    // نحتاج إلى استيراد مكتبة userdata للوصول إلى الأسرار في Colab
    // هذا ليس جزءًا من Go القياسي، بل خاص ببيئة Colab Python
    // لذلك، سنقرأ المفتاح من متغير بيئة يتم تعيينه لاحقًا
)

func main() {
	// قراءة مفتاح API من متغير البيئة
    // سيتم تعيين هذا المتغير في خطوة التشغيل التالية
	apiKey := os.Getenv("GEMINI_API_KEY_ENV")
	if apiKey == "" {
		log.Fatal("لم يتم تعيين متغير البيئة GEMINI_API_KEY_ENV")
	}

	ctx := context.Background()
	// إنشاء عميل Gemini
	client, err := genai.NewClient(ctx, option.WithAPIKey(apiKey))
	if err != nil {
		log.Fatalf("فشل في إنشاء العميل: %v", err)
	}
	defer client.Close()

	// اختيار نموذج Gemini (gemini-1.5-flash هو خيار جيد وسريع ومجاني للاستخدامات الأساسية)
	model := client.GenerativeModel("gemini-1.5-flash-latest")
    // يمكنك أيضًا تجربة "gemini-pro"

	fmt.Println("مرحباً! أنا شات بوت بسيط. اكتب 'خروج' للإنهاء.")
	fmt.Println("----------------------------------------------")

	// بدء حلقة المحادثة
	scanner := bufio.NewScanner(os.Stdin)
	for {
		fmt.Print("أنت: ")
		// قراءة إدخال المستخدم
		if !scanner.Scan() {
			if err := scanner.Err(); err != nil {
				log.Printf("خطأ في قراءة الإدخال: %v", err)
			}
			break // الخروج إذا كان هناك خطأ أو نهاية الإدخال
		}
		userInput := scanner.Text()

		// التحقق من أمر الخروج
		if strings.ToLower(userInput) == "خروج" {
			fmt.Println("وداعاً!")
			break
		}

		// إرسال الإدخال إلى نموذج Gemini
		resp, err := model.GenerateContent(ctx, genai.Text(userInput))
		if err != nil {
			log.Printf("فشل في إنشاء الاستجابة: %v", err)
            fmt.Println("بوت: عذراً، حدث خطأ أثناء معالجة طلبك.")
			continue // الاستمرار في الحلقة التالية
		}

		// طباعة استجابة النموذج
		printResponse(resp)
	}
}

// دالة مساعدة لطباعة الاستجابة بشكل منظم
func printResponse(resp *genai.GenerateContentResponse) {
	fmt.Print("بوت: ")
	if resp == nil || len(resp.Candidates) == 0 || resp.Candidates[0].Content == nil || len(resp.Candidates[0].Content.Parts) == 0 {
		fmt.Println("لم يتم تلقي أي محتوى من النموذج.")
		return
	}
	// استخراج النص من الاستجابة
	// قد تكون الاستجابة مقسمة لأجزاء، نجمعها هنا
	var responseText strings.Builder
	for _, part := range resp.Candidates[0].Content.Parts {
        // نتأكد أن الجزء هو نص قبل إضافته
        if textPart, ok := part.(genai.Text); ok {
             responseText.WriteString(string(textPart))
        }
	}
    fmt.Println(responseText.String())
}
```

```python
%%writefile chatbot.go
package main

import (
	"bufio"
	"bytes" // لاستخدام bytes.Buffer
	"context"
	"encoding/json" // لمعالجة JSON
	"fmt"
	"io"      // لقراءة جسم الاستجابة
	"log"
	"net/http" // لعمل طلبات HTTP
	"os"
	"strings"
	"time" // لإضافة تأخير بسيط إذا لزم الأمر
)

// تعريف هيكل لطلب JSON إلى Hugging Face API
type HfRequestPayload struct {
	Inputs     string            `json:"inputs"`
	Parameters map[string]interface{} `json:"parameters,omitempty"` // بارامترات اختيارية
	Options    map[string]interface{} `json:"options,omitempty"`    // خيارات مثل wait_for_model
}

// تعريف هيكل لاستقبال الاستجابة JSON (قد يختلف قليلاً حسب النموذج)
// غالبًا ما يكون الاستدلال النصي في مصفوفة
type HfResponsePayload []struct {
	GeneratedText string `json:"generated_text"`
}

// --- المتغيرات العامة ---
var (
	huggingFaceApiToken string
	httpClient          *http.Client
	modelUrl            string
)

// اسم النموذج المطلوب
const modelID = "meta-llama/Llama-3.1-8B-Instruct" // <-- ملاحظة: استخدام 8B Instruct لأنه متاح بشكل أوسع حاليًا.
// جرب "meta-llama/Llama-3.2-3B-Instruct" ولكن قد لا يكون متاحًا أو يتطلب وصولاً خاصًا على Inference API المجانية

func main() {
	// قراءة توكن Hugging Face من متغير البيئة
	huggingFaceApiToken = os.Getenv("HF_API_TOKEN_ENV")
	if huggingFaceApiToken == "" {
		log.Fatal("لم يتم تعيين متغير البيئة HF_API_TOKEN_ENV")
	}

	// إعداد عميل HTTP
	httpClient = &http.Client{Timeout: 60 * time.Second} // زيادة المهلة للنماذج الكبيرة

	// بناء عنوان URL للنموذج
	modelUrl = fmt.Sprintf("https://api-inference.huggingface.co/models/%s", modelID)
	log.Printf("استخدام النموذج: %s عبر URL: %s\n", modelID, modelUrl)


	fmt.Println("مرحباً! أنا شات بوت يتحدث مع Hugging Face. اكتب 'خروج' للإنهاء.")
    fmt.Printf("النموذج المستخدم: %s (قد يستغرق الرد الأول بعض الوقت لتحميل النموذج على الخادم)\n", modelID)
	fmt.Println("---------------------------------------------------------------------------------")

	// بدء حلقة المحادثة
	scanner := bufio.NewScanner(os.Stdin)
	for {
		fmt.Print("أنت: ")
		if !scanner.Scan() {
			if err := scanner.Err(); err != nil {
				log.Printf("خطأ في قراءة الإدخال: %v", err)
			}
			break
		}
		userInput := scanner.Text()

		if strings.ToLower(userInput) == "خروج" {
			fmt.Println("وداعاً!")
			break
		}

		// استدعاء Hugging Face API
		botResponse, err := queryHuggingFaceModel(userInput)
		if err != nil {
			log.Printf("فشل في الحصول على استجابة من Hugging Face: %v", err)
			fmt.Println("بوت: عذراً، حدث خطأ أثناء معالجة طلبك.")
			// قد يكون النموذج لا يزال قيد التحميل، خاصة في المرة الأولى
            if strings.Contains(err.Error(), "503") {
                 fmt.Println("بوت: (قد يكون النموذج لا يزال قيد التحميل، يرجى المحاولة مرة أخرى بعد قليل)")
            }
			continue
		}

		// طباعة استجابة البوت
		fmt.Println("بوت:", botResponse)
	}
}

// دالة لإرسال الطلب إلى Hugging Face API والحصول على الرد
func queryHuggingFaceModel(prompt string) (string, error) {
	// إعداد بيانات الطلب
	payload := HfRequestPayload{
		Inputs: prompt,
        // يمكنك إضافة بارامترات هنا إذا أردت
        // Parameters: map[string]interface{}{
        //     "max_new_tokens": 100,
        //     "temperature": 0.7,
        // },
        // انتظر حتى يتم تحميل النموذج إذا لم يكن جاهزًا
        Options: map[string]interface{}{
            "wait_for_model": true,
            "use_cache": false, // تعطيل الكاش للحصول على ردود جديدة دائمًا (اختياري)
        },
	}

	// تحويل البيانات إلى JSON
	jsonData, err := json.Marshal(payload)
	if err != nil {
		return "", fmt.Errorf("فشل في تحويل الطلب إلى JSON: %w", err)
	}

	// إنشاء طلب HTTP POST
	req, err := http.NewRequestWithContext(context.Background(), "POST", modelUrl, bytes.NewBuffer(jsonData))
	if err != nil {
		return "", fmt.Errorf("فشل في إنشاء طلب HTTP: %w", err)
	}

	// إضافة الترويسات اللازمة (التوثيق ونوع المحتوى)
	req.Header.Set("Authorization", "Bearer "+huggingFaceApiToken)
	req.Header.Set("Content-Type", "application/json")

	// إرسال الطلب
	resp, err := httpClient.Do(req)
	if err != nil {
		return "", fmt.Errorf("فشل في إرسال طلب HTTP: %w", err)
	}
	defer resp.Body.Close()

	// قراءة جسم الاستجابة
	bodyBytes, err := io.ReadAll(resp.Body)
	if err != nil {
		return "", fmt.Errorf("فشل في قراءة جسم الاستجابة: %w", err)
	}

    log.Printf("كود حالة الاستجابة: %d", resp.StatusCode)
    // log.Printf("جسم الاستجابة الخام: %s", string(bodyBytes)) // للتحقق من الاستجابة الكاملة إذا لزم الأمر


	// التحقق من كود حالة HTTP
	if resp.StatusCode != http.StatusOK {
        // محاولة فك تشفير رسالة الخطأ من HF إذا كانت JSON
        var errorResp map[string]interface{}
        if json.Unmarshal(bodyBytes, &errorResp) == nil {
             if errMsg, ok := errorResp["error"]; ok {
                 return "", fmt.Errorf("خطأ من API (الحالة %d): %v", resp.StatusCode, errMsg)
             }
        }
		// إذا لم تكن JSON أو لم تحتوِ على حقل الخطأ
		return "", fmt.Errorf("خطأ من API: كود الحالة %d, الجسم: %s", resp.StatusCode, string(bodyBytes))
	}

	// فك تشفير استجابة JSON
	var responsePayload HfResponsePayload
	err = json.Unmarshal(bodyBytes, &responsePayload)
	if err != nil {
		// قد لا تكون الاستجابة JSON المتوقع، حاول طباعة الجسم مباشرة
		log.Printf("فشل في فك تشفير JSON: %v. الجسم الخام: %s", err, string(bodyBytes))
        // بعض النماذج قد ترجع نصًا عاديًا أو بنية مختلفة
        // كحل بديل بسيط، حاول إرجاع الجسم كما هو إذا لم يكن فارغًا
        if len(bodyBytes) > 0 {
            return string(bodyBytes), nil // قد يحتاج لتنظيف إضافي
        }
		return "", fmt.Errorf("فشل في فك تشفير استجابة JSON: %w", err)
	}

	// التحقق من وجود بيانات في الاستجابة
	if len(responsePayload) == 0 || responsePayload[0].GeneratedText == "" {
        log.Printf("الاستجابة JSON تم فكها ولكنها فارغة أو لا تحتوي على generated_text. الجسم الخام: %s", string(bodyBytes))
		return "", fmt.Errorf("لم يتم العثور على 'generated_text' في استجابة API")
	}

	// استخراج النص المُنشأ
    // غالبًا ما تتضمن استجابة HF النص الأصلي + النص الجديد، نحتاج لتنظيفه أحيانًا
    generatedText := responsePayload[0].GeneratedText
    // إزالة النص الأصلي إذا كان مضمنًا في البداية (هذا يعتمد على سلوك النموذج)
    if strings.HasPrefix(generatedText, prompt) {
        cleanedText := strings.TrimPrefix(generatedText, prompt)
        // إزالة أي مسافات بادئة أو أسطر جديدة قد تظهر بعد الإزالة
        return strings.TrimSpace(cleanedText), nil
    }

	return generatedText, nil
}
```

```python
// %%writefile chatbot_local.go  <-- هذا لن يعمل كما هو

package main

import (
	"bufio"
	"fmt"
	"log"
	"os"
	"strings"

	// مثال على مكتبة روابط - قد تحتاج لاستخدام مكتبة مختلفة
	llama "github.com/go-skynet/go-llama.cpp" // أو أي مكتبة روابط أخرى تجدها
)

func main() {
	// --- هذه الخطوات تتطلب إعدادًا مسبقًا معقدًا خارج هذا الكود ---
	modelPath := "/content/models/llama-3.2-3b-instruct.Q4_K_M.gguf" // <--- مثال لمسار النموذج الذي يجب تنزيله ووضعه هنا

	// تحقق من وجود النموذج (خطوة أساسية)
	if _, err := os.Stat(modelPath); os.IsNotExist(err) {
		log.Fatalf("ملف النموذج غير موجود في المسار: %s. يجب تنزيله أولاً ووضعه في المسار الصحيح.", modelPath)
	}
	// --- نهاية الإعداد المسبق ---


	// إعداد خيارات النموذج (تعتمد على مكتبة الروابط المستخدمة)
	opts := []llama.llamaOption{
		llama.SetModel(modelPath),
		llama.SetContext(512), // حجم السياق
		// قد تحتاج لخيارات أخرى مثل استخدام GPU إذا تم بناء llama.cpp بدعمه وتم توفيره في Colab
		// llama.EnableF16Memory(),
		// llama.SetGPULayers(20), // عدد الطبقات لوضعها على GPU
	}

	fmt.Println("محاولة تحميل النموذج المحلي (قد يستغرق وقتًا طويلاً أو يفشل بسبب الذاكرة)...")
	l, err := llama.New(opts...) // *** هذه هي النقطة التي ستفشل غالبًا في Colab ***
	if err != nil {
		log.Fatalf("فشل في تحميل النموذج من %s: %v\nقد يكون السبب نقص الذاكرة في Colab.", modelPath, err)
	}
	defer l.Close() // تأكد من إغلاق النموذج لتحرير الموارد (إذا تم تحميله بنجاح)
	fmt.Println("تم تحميل النموذج بنجاح (إذا وصلت إلى هنا!).")


	fmt.Println("مرحباً! أنا شات بوت يعمل محليًا (نظريًا). اكتب 'خروج' للإنهاء.")
	fmt.Println("-----------------------------------------------------------------")

	scanner := bufio.NewScanner(os.Stdin)
	for {
		fmt.Print("أنت: ")
		if !scanner.Scan() {
			break
		}
		userInput := scanner.Text()

		if strings.ToLower(userInput) == "خروج" {
			fmt.Println("وداعاً!")
			break
		}

		// تنسيق الإدخال قد يكون ضروريًا لنموذج Instruct
		// prompt := fmt.Sprintf("[INST] %s [/INST]", userInput) // مثال لتنسيق Llama Instruct

		// تشغيل الاستدلال (يعتمد على واجهة المكتبة)
		fmt.Print("بوت: ")
		_, err := l.Predict(userInput, // أو prompt
            // خيارات الاستدلال (تعتمد على المكتبة)
            llama.SetTokenCallback(func(token string) bool {
                fmt.Print(token) // طباعة التوكنات فور توليدها
                return true
            }),
            llama.SetTokens(200), // الحد الأقصى للتوكنات الجديدة
            llama.SetThreads(4), // عدد خيوط المعالجة
            // ... خيارات أخرى مثل temperature, top_p, top_k
        )

		fmt.Println() // لسطر جديد بعد الرد

		if err != nil {
			log.Printf("حدث خطأ أثناء الاستدلال: %v", err)
            // قد يكون خطأ ذاكرة آخر هنا أيضًا
		}
	}
}
```

```python
%%writefile chatbot_local.go
package main

import (
	"bufio"
	"fmt"
	"log"
	"os"
	"strings"
	"runtime" // للحصول على عدد المعالجات

	llama "github.com/go-skynet/go-llama.cpp"
)

func main() {
	// قراءة مسار النموذج من متغير البيئة الذي تم تعيينه سابقًا
	modelPath := os.Getenv("MODEL_PATH")
	if modelPath == "" {
		// إذا لم يتم تعيين متغير البيئة، استخدم قيمة افتراضية (أو أوقف البرنامج)
		// تأكد من أن هذه القيمة تطابق النموذج الذي قمت بتنزيله بالفعل!
		modelPath = "/content/models/llama-3.2-3b-instruct.Q4_K_M.gguf" // <--- تأكد من صحة هذا المسار
		log.Printf("تحذير: متغير البيئة MODEL_PATH غير معين، استخدام القيمة الافتراضية: %s", modelPath)
	}

	// التحقق من وجود ملف النموذج
	if _, err := os.Stat(modelPath); os.IsNotExist(err) {
		log.Fatalf("خطأ فادح: ملف النموذج غير موجود في المسار المحدد: %s. تأكد من تنزيله وتعيين المسار الصحيح.", modelPath)
	}

	// إعداد خيارات النموذج
	// استخدم عدد معالجات منطقية أقل بواحد لتجنب استهلاك كل الموارد
	numThreads := runtime.NumCPU() - 1
	if numThreads < 1 {
		numThreads = 1
	}

	opts := []llama.llamaOption{
		llama.SetModel(modelPath),
		llama.SetContext(512),     // حجم السياق (يمكن زيادته إذا سمحت الذاكرة)
		llama.SetThreads(numThreads), // عدد خيوط المعالجة
		// llama.SetGPULayers(10), // <-- لا تستخدم هذا إلا إذا بنيت llama.cpp بدعم GPU ولديك GPU متاح!
	}

	fmt.Printf("محاولة تحميل النموذج المحلي من: %s\n", modelPath)
	fmt.Println("!!!!!!!!! تحذير: هذه الخطوة قد تستغرق وقتًا طويلاً جداً وقد تفشل بسبب نفاد ذاكرة RAM في Colab !!!!!!!!!")

	l, err := llama.New(opts...)
	if err != nil {
		// *** هذه هي نقطة الفشل الأكثر احتمالاً ***
		log.Fatalf("!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!\n"+
			"           فشل فادح في تحميل النموذج: %v\n"+
			"           السبب الأكثر احتمالاً هو نفاد ذاكرة الوصول العشوائي (RAM) في Colab.\n"+
			"           جرب نموذجًا أصغر (مثل 3B Q4_K_M)، أو استخدم Colab Pro بذاكرة أكبر،\n"+
			"           أو قم بالتشغيل على جهاز محلي بموارد كافية.\n"+
			"           الحل الموصى به في Colab هو استخدام واجهات برمجة التطبيقات (APIs).\n"+
			"!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!", err)
	}
	defer l.Close() // تأكد من إغلاق النموذج لتحرير الموارد

	fmt.Println("*************************************************")
	fmt.Println("**** تم تحميل النموذج بنجاح! (هذا إنجاز!) ****")
	fmt.Println("*************************************************")
	fmt.Println("مرحباً! أنا شات بوت يعمل محليًا. اكتب 'خروج' للإنهاء.")
	fmt.Println("-----------------------------------------------------------------")

	scanner := bufio.NewScanner(os.Stdin)
	var chatHistory strings.Builder // لتخزين سجل بسيط للمحادثة (اختياري)

	for {
		fmt.Print("أنت: ")
		if !scanner.Scan() {
			break
		}
		userInput := scanner.Text()

		if strings.ToLower(userInput) == "خروج" {
			fmt.Println("وداعاً!")
			break
		}

		// بناء الـ prompt (قد تحتاج لتعديله ليناسب نموذج Instruct المحدد)
		// هذا مثال بسيط، قد تحتاج لتنسيق أكثر تعقيدًا مع أدوار المستخدم/المساعد
		// currentPrompt := chatHistory.String() + "أنت: " + userInput + "\nبوت: "
		// استخدام تنسيق Llama 3 Instruct الأساسي
		currentPrompt := fmt.Sprintf("<|begin_of_text|><|start_header_id|>user<|end_header_id|>\n\n%s<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n\n", userInput)


		fmt.Print("بوت: ")
		// تشغيل الاستدلال
		_, err := l.Predict(currentPrompt,
			llama.SetTokenCallback(func(token string) bool {
				fmt.Print(token) // طباعة التوكنات فور توليدها
				return true      // استمر في التوليد
			}),
			llama.SetTokens(256),       // الحد الأقصى للتوكنات الجديدة
			llama.SetThreads(numThreads), // إعادة تحديد عدد الخيوط (قد لا يكون ضروريًا)
			llama.SetTopK(40),          // Top-K sampling
			llama.SetTopP(0.95),        // Top-P (Nucleus) sampling
			llama.SetTemperature(0.7),  // درجة الحرارة (للتحكم في العشوائية)
			llama.SetStop("\nأنت:", "<|eot_id|>"), // كلمات التوقف
		)

		fmt.Println() // لسطر جديد بعد الرد الكامل

		if err != nil {
			log.Printf("حدث خطأ أثناء الاستدلال: %v", err)
			// قد يحدث خطأ ذاكرة هنا أيضًا أثناء التوليد
		} else {
			// إضافة التفاعل الأخير إلى السجل (إذا نجح)
            // تحتاج إلى طريقة لالتقاط الرد الكامل لتحديث السجل بشكل صحيح
            // هذه الطريقة معقدة مع callback، قد تحتاج لتخزين الرد في متغير خارجي
            // chatHistory.WriteString("أنت: " + userInput + "\n")
            // chatHistory.WriteString("بوت: " + "الرد الكامل هنا" + "\n") // <-- كيف تحصل على الرد الكامل؟
		}
	}
}
```

```python
import os

def find_file(filename, root_dir):
  """
  هذه الدالة تبحث عن ملف باسم محدد في دليل محدد وجميع أدلته الفرعية.

  Args:
    filename: اسم الملف المراد البحث عنه.
    root_dir: الدليل الرئيسي للبدء بالبحث فيه.

  Returns:
    المسار الكامل للملف إذا تم العثور عليه، وإلا None.
  """

  for root, dirs, files in os.walk(root_dir):
    if filename in files:
      return os.path.join(root, filename)
  return None

# مثال على الاستخدام
filename = "libllama.so"
root_dir = "/content/"
file_path = find_file(filename, root_dir)

if file_path:
  print("تم العثور على الملف في:", file_path)
else:
  print("لم يتم العثور على الملف.")
```
